```python
%sql
CREATE OR REPLACE TABLE sample_claims_dev.03_gold.customer_claim_policy_telematics_predicted_gold AS
SELECT t.*,a.* EXCEPT(chassis_no,claim_no) FROM sample_claims_dev.03_gold.customer_claim_policy_telematics_gold t
JOIN sample_claims_dev.03_gold.claim_images_predicted_gold a USING (claim_no)
```

```python
%sql
CREATE OR REPLACE TABLE sample_claims_dev.02_silver.claims_rules_silver (rule_id BIGINT, rule STRING, check_name STRING,check_code STRING,check_severity STRING, is_active BOOLEAN);
```

```python
def insert_rule(rule,name,code,severity,is_active):
    spark.sql(f"INSERT into sample_claims_dev.02_silver.claims_rules_silver(rule,check_name,check_code,check_severity,is_active) values('{rule}','{name}','{code}','{severity}','{is_active}')")
```

```python
invalid_policy_date = '''
  CASE WHEN to_date(pol_eff_date, "yyyy-MM-dd") < to_date(claim_date) and to_date(pol_expiry_date, "yyyy-MM-dd") < to_date(claim_date) THEN "VALID" 
  ELSE "NOT VALID"  
  END
'''

insert_rule('invalid policy date', 'valid_date', invalid_policy_date, 'HIGH', True)
```

```python
exceeds_policy_amount = '''
CASE WHEN  sum_insured >= total 
    THEN "claim value in the range of premium"
    ELSE "claim value more than premium"
END 
'''
insert_rule('exceeds policy amount', 'valid_amount', exceeds_policy_amount, 'HIGH', True)
```

```python
severity_mismatch = '''
CASE WHEN    incident_severity="Total Loss" AND damage_prediction.label = "major" THEN  "Severity matches the report"
       WHEN  incident_severity="Major Damage" AND damage_prediction.label = "minor"  THEN  "Severity matches the report"
       WHEN  (incident_severity="Minor Damage" OR incident_severity="Trivial Damage") AND damage_prediction.label = "ok" THEN  "Severity matches the report"
       ELSE "Severity does not match"
END 
'''

insert_rule('severity mismatch', 'reported_severity_check', severity_mismatch, 'HIGH', True)
```

```python
exceeds_speed = '''
CASE WHEN  telematics_speed <= 45 and telematics_speed > 0 THEN  "Normal Speed"
       WHEN telematics_speed > 45 THEN  "High Speed"
       ELSE "Invalid speed"
END
'''
insert_rule('exceeds speed', 'speed_check', exceeds_speed, 'HIGH', True)
```

```python
release_funds = '''
CASE WHEN  reported_severity_check="Severity matches the report" and valid_amount="claim value in the range of premium" and valid_date="VALID" then "release funds"
       ELSE "claim needs more investigation" 
END
'''
insert_rule('release funds', 'fund_release', release_funds, 'HIGH', True)
```

```python
from pyspark.sql.functions import expr

df = spark.sql("SELECT * FROM sample_claims_dev.03_gold.customer_claim_policy_telematics_predicted_gold")
rules = spark.sql('SELECT * FROM sample_claims_dev.02_silver.claims_rules_silver where is_active=true order by rule_id').collect()

# Separate rules into base checks and dependent checks
base_rules = [r for r in rules if r.check_name != 'fund_release']
dependent_rules = [r for r in rules if r.check_name == 'fund_release']

# Apply base rules first
for rule in base_rules:
  print(rule.rule, rule.check_code)
  # Fix column name mismatches
  corrected_code = rule.check_code.replace('incident_severity', 'severity')
  corrected_code = corrected_code.replace('pol_eff_date', 'POL_EFF_DATE')
  corrected_code = corrected_code.replace('pol_expiry_date', 'POL_EXPIRY_DATE')
  corrected_code = corrected_code.replace('sum_insured', 'SUM_INSURED')
  df=df.withColumn(rule.check_name, expr(corrected_code))

# Apply dependent rules (that reference columns created by base rules)
for rule in dependent_rules:
  print(rule.rule, rule.check_code)
  df=df.withColumn(rule.check_name, expr(rule.check_code))

#overwrite table with new insights
df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("sample_claims_dev.03_gold.claim_insights_gold")
```
